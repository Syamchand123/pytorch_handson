**working on building neural network form scrath**


````python
class Myneuralnetwork():
  def __init__(self,x):
    torch.manual_seed(100)
    self.weights = torch.rand(x.shape[1],dtype=torch.float64,requires_grad=True)
    self.bias = torch.zeros(1,dtype=torch.float64,requires_grad=True)

  def forward(self,x):
    z=torch.matmul(x,self.weights)+self.bias
    y_pred=torch.sigmoid(z)
    return y_pred


  def calculateloss(self,y_pred,y):
    ep=1e-7
    y_pred=torch.clamp(y_pred,ep,1-ep)
    loss=-(y_train_tensor * torch.log(y_pred) + (1-y_train_tensor) *torch.log(1-y_pred)).mean()
    return loss
````

````python
#parameters
learning_rate=0.1
epochs=10
````

**pipeline**


````python
#create model
model=Myneuralnetwork(x_train_tensor)


for epoch in range(epochs):
 #forward pass
 y_pred=model.forward(x_train_tensor)

 #calculate loss
 Loss=model.calculateloss(y_pred,y_train_tensor)

 #backward propagation
 Loss.backward()

 #updating parameters
 with torch.no_grad():
  model.weights -= learning_rate *model.weights.grad
  model.bias -= learning_rate*model.bias.grad

 #zero grads
 model.weights.grad.zero_()
 model.bias.grad.zero_()


 #print loss in each
 print(f'Epoch: {epoch+1} , Loss : {Loss.item()}')
````

````python
import torch
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
````

````python
df=pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/master/data.csv')
df.head()
````

````python
df.drop(columns=['id','Unnamed: 32'],inplace=True)
````

````python
df.shape
````

````python
x_train,x_test,y_train,y_test=train_test_split(df.iloc[:,1:],df.iloc[:,0],test_size=0.2)
````

````python
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
````

````python
encode=LabelEncoder()
y_train=encode.fit_transform(y_train)
y_test=encode.fit_transform(y_test)
````

````python
x_train_tensor=torch.from_numpy(x_train)
x_test_tensor=torch.from_numpy(x_test)
y_train_tensor=torch.from_numpy(y_train)
y_test_tensor=torch.from_numpy(y_test)
````

````python
#parameters
learning_rate=0.1
epochs=200
````

````python
#design model
class Myneuralnet(nn.Module):

  def __init__(self,num_features):
    super().__init__()
    self.net=nn.Sequential(
        nn.Linear(num_features,3),
        nn.ReLU(),
        nn.Linear(3,1),
        nn.Sigmoid()
    )

  def forward(self,features):
    out=self.net(features)
    return out

````

````python
#instance model and required funtions
model=Myneuralnet(x_train_tensor.shape[1])
model = model.double()
Loss_funtion=nn.BCELoss()
optimizer=torch.optim.SGD(model.parameters(),lr=learning_rate)

for epoch in range(epochs):
  # call model
  y_pred=model(x_train_tensor)


  #calculate loss
  Loss=Loss_funtion(y_pred,y_train_tensor.view(-1,1).double())

  #clear gradients
  optimizer.zero_grad()


  #backward propagation
  Loss.backward()


  #optimizer
  optimizer.step()


  #print loss in each
  print(f'Epoch: {epoch+1} , Loss : {Loss.item()}')


````

````python
#model evaluation
with torch.no_grad():
  y_pred_test=model(x_test_tensor)
  y_pred_test=(y_pred_test > 0.5).float()
  # Ensure y_test_tensor matches the shape and type of y_pred_test for explicit comparison
  y_test_tensor_formatted = y_test_tensor.float().unsqueeze(1)
  accuracy=(y_pred_test == y_test_tensor_formatted).float().mean()
  print(accuracy.item())
````
